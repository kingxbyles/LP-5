```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
import tensorflow as tf
```

```python
data = pd.read_csv("../IMDB Dataset.csv")
data.head()
```

```python
X = data['review']
y = data['sentiment']
```

```python
y = y.map({
    'positive':1,
    'negative':0
})
```

```python
vectorizer = TfidfVectorizer(max_features = 10000)

X = vectorizer.fit_transform(X).toarray()
```

```python
reviews = data['review']

X_train, X_test, y_train, y_test, reviews_train, reviews_test = train_test_split(X, y, reviews, test_size = 0.2)
```

```python
model = tf.keras.Sequential([

    tf.keras.layers.Dense(
        64,
        activation = 'relu',
        input_shape = (X_train.shape[1],)
    ),

    tf.keras.layers.Dense(
        32,
        activation = 'relu',
    ),

    tf.keras.layers.Dense(
        1,
        activation = 'sigmoid',
    )
    
])
```

```python
model.compile(
    optimizer = 'adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(X_train, y_train, epochs = 5)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)

print("Avg Loss: ", loss*100)
print("Avg Accuracy: ", accuracy*100)
```

```python
sample = reviews_test.iloc[4]
print("Sample Review: ", sample)

prediction = model.predict(X_test[4:5])

if prediction[0][0] > 0.5:
    print("\nPredicted Sentiment: Positive")
else:
    print("\nPredicted Sentiment: Negative")
```
